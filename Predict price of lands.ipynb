```python
outlier_percentage = {}
for feature in ['investment_amount', 'area', 'distance_to_highway', "starting_price", "final_price_TARGET" ]:
    tempData = data.sort_values(by=feature)[feature]
    Q1, Q3 = tempData.quantile([0.25, 0.75])
    IQR = Q3 - Q1
    Lower_range = Q1 - (1.5 * IQR)
    Upper_range = Q3 + (1.5 * IQR)
    outlier_percentage[feature] = round((((tempData<(Q1 - 1.5 * IQR)) | (tempData>(Q3 + 1.5 * IQR))).sum()/tempData.shape[0])*100,2)
outlier_percentage
```

```python
outlier = data[(data[feature]>Lower_range) & (data[feature]<Upper_range)].reset_index(drop=True)


outlier.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state = 0)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#from sklearn.preprocessing import StandardScaler
#sc = StandardScaler()
#x_train = sc.fit_transform(x_train)
#x_test = sc.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
model1 = RandomForestRegressor()
model1.fit(x_train, y_train)
y_pred1 = model1.predict(x_test)
from sklearn.metrics import r2_score, mean_squared_error
mse = mean_squared_error(y_test, y_pred1)
rmse = np.sqrt(mse)
print("RMSE score: ", rmse)

r2_score=r2_score(y_test, y_pred1)
print("R2 score: ", r2_score)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
model2 = GradientBoostingRegressor()
model2.fit(x_train, y_train)
y_pred2 = model2.predict(x_test)
from sklearn.metrics import r2_score, mean_squared_error
mse = mean_squared_error(y_test, y_pred2)
rmse = np.sqrt(mse)
print("RMSE score: ", rmse)

r2_score=r2_score(y_test, y_pred2)
print("R2 score: ", r2_score)
```
